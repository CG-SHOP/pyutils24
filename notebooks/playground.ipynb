```python
import random

from cgshop_pyutils24 import Coordinate, Knapsack, QuadTree, Rectangle
```

```python
a = Rectangle(0, 0, 10, 10)
assert a.volume() == 100
```

```python
b = Rectangle(0, 0, 10, 20)
assert b.volume() == 200
```

```python
assert a.do_overlap(b) is True
```

```python
c = Rectangle(0, 10, 10, 10)
assert a.do_overlap(c) is False
assert b.do_overlap(c) is True
```

```python
qt = QuadTree(Rectangle(0, 0, 100, 100))
```

```python
qt.add(a, 0)
qt.add(b, 1)
qt.add(c, 2)
```

```python
result = qt.query(Rectangle(0, 0, 100, 100))
result
```

```python
result = qt.query(Rectangle(0, 0, 10, 10))
result
```

```python
qt2 = QuadTree(Rectangle(0, 0, 10_000, 10_000))
for i in range(100_000):
    x = random.randint(0, 10_000 - 10)
    y = random.randint(0, 10_000 - 10)
    w = random.randint(1, 10)
    h = random.randint(1, 10)
    qt2.add(Rectangle(x, y, w, h), i)
print(qt2.depth())
```

```python
print(len(qt2))
```

```python
qt2.query(Rectangle(0, 0, 1_000, 1_000))
```
